```python
from __future__ import print_function, unicode_literals
import sys
sys.path.append("../")
import jieba
jieba.load_userdict("/content/user_dict.txt")
import jieba.posseg as pseg
```

```python
with open('/content/sample.txt', 'r', encoding='utf-8') as file:
    text = file.read()
seg_list = jieba.cut_for_search(text)
print(', '.join(seg_list))
```

```python
seg_list = jieba.cut_for_search(text)
seg_list = list(seg_list)

def contains_street_name(segment):
    street_keywords = ['街', '路', '道', '巷']
    for keyword in street_keywords:
        if keyword in segment:
            return True
    return False

street_segments = [(i, segment) for i, segment in enumerate(seg_list) if contains_street_name(segment)]
for index, segment in street_segments:
    print(f"Index: {index}, Segment: {segment}")
```
